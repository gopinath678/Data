```python
def fetch_html(web_url):
    import urllib2
    html_page = urllib2.urlopen(web_url).read()
    print html_page
    text = html2text(html_page)
    return text
```

```python
def html2text(html_page):
    from bs4 import BeautifulSoup
    text = ""
    soup = BeautifulSoup(html_page, 'html.parser')
    for script in soup.find_all("script"):
        soup.script.extract()
    for para in soup.find_all("p"):
        text = text + "\n\n" + para.get_text()
    return text
```

```python
text = fetch_html("http://www.livesinabox.com/friends/season1/108ndt.htm")
out_file = open("season1/episode-8.txt", "w")
out_file.write(text)
```
